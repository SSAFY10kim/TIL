## C. 데이터 분석 - 2021년 이후 최고, 최저 종가 출력

• csv 파일을 DataFrame 으로 읽어와 2021년 이후의 데이터만 필터링합니다.

• 종가(Close) 필드를 활용하여, 2021년 이후 가장 높은 종가와 가장 낮은 종가를 출력

• Pandas 의 내장 함수를 사용합니다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path = "netflix_data/NFLX.csv"

df = pd.read_csv(csv_path, usecols = range(0,5))

df
```

```python
df_after_2021 = df[df["Date"] >= "2021-01-01"]

df_after_2021
```

```python
df_after_2021.loc[:, "Close"]
```

```python
max_price = df_after_2021["Close"].max()
min_price = df_after_2021["Close"].min()

print("최고 종가: ", max_price)
print("최저 종가: ", min_price)
```
